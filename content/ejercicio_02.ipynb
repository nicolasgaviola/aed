```python
""" Escritura de datos de un csv (Página 36)
Escritura de datos de un csv a partir de una estructura de datos en Python
"""
```

```python
import csv
import os
```

```python
ruta = os.path.join(os.getcwd(), "nuevos_datos.csv")
```

```python
with open(ruta, mode="w", newline="", encoding="utf-8") as archivo:
    escritor = csv.writer(archivo)
    escritor.writerow(["Nombre", "Edad", "Ciudad"])
    escritor.writerow(["Juan Pérez", 30, "Formosa"])
    escritor.writerow(["Maria Gomez", 25, "Clorinda"])
    escritor.writerow(["Luis Rodriguez", 35, "Formosa"])
    escritor.writerow(["Lucas Rodriguez", 38, "El Colorado"])
    escritor.writerow(["Ana Martinez", 28, "Formosa"])
```

```python
print("Datos escritos correctamente en el archivo 'nuevos_datos.csv'")
```
